Now let's get rid ouf the outliers with a "3-sigma" method (keeping only the values that lies within [m-3s, m+3s])

```python
def remove_outliers(df):
    df = df.copy()
    outliers = {}
    for col in df.columns:
        if str(df[col].dtype) != "category":
            df = df[np.abs(df[col]-df[col].mean()) < (3*df[col].std())]
            outliers = df[~(np.abs(df[col]-df[col].mean()) < (3*df[col].std()))]
    return df, outliers
```

###### 1. Functions to compare the models

```python
def show_metrics(y_true, y_score):
    
    """ simple function that outputs the classification metrics"""
    
    # True positive
    tp = np.sum(y_true * y_score)
    # False positive
    fp = np.sum((y_true == 0) * y_score)
    # True negative
    tn = np.sum((y_true == 0) * (y_score==0))
    # False negative
    fn = np.sum(y_true * (y_score==0))

    # True positive rate (sensitivity or recall)
    tpr = tp / (tp + fn)
    # False positive rate (fall-out)
    fpr = fp / (fp + tn)
    # Precision
    precision = tp / (tp + fp)
    # True negatvie tate (specificity)
    tnr = 1 - fpr
    # F1 score
    f1 = 2*tp / (2*tp + fp + fn)
    # ROC-AUC for binary classification
    auc = roc_auc_score(y_true, y_score)
    
    print(colored('True positive: %.2f' % tp, 'green'))
    print(colored('False positive: %.2f' % fp, 'red'))
    print(colored("True negative: %.2f" % tn, 'green'))
    print(colored("False negative: %.2f" % fn, 'red'))
    
    print("True positive rate (recall): %.3f" % tpr)
    print("False positive rate: %.3f" % fpr)
    print("True negative rate: %.3f" % tnr)
    
    print("Precision: %.3f" % precision)
    print("F1: %.3f" % f1)
    print("ROC-AUC: %.3f" % auc)
```

```python
def show_feature_importance(clf, X, feature_names):
    
    """ plots the feature importances according to scikit-learn's feature_importances_ computation """
    
    importances = clf.feature_importances_
    std = np.std([tree.feature_importances_ for tree in clf.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1][:50]

    # Print the feature ranking
    print("Feature ranking:")

    for f in range(50):
        print("%d. feature %s (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))

    # Plot the feature importances of the forest
    plt.figure(figsize=(16,8))
    plt.title("Feature importances")
    plt.bar(range(50), importances[indices],
           color="r", yerr=std[indices], align="center")
    plt.xticks(range(50), indices)
    plt.xlim([-1, 50])
    plt.show()
```

```python
def show_feature_contribution(clf, X):
    
    """ plots the feature contribution according to treeinterpreter """
    
    nb = X.shape[0]
    X = X.iloc[:nb,:] 
    
    prediction, bias, contributions = ti.predict(clf, X)
    
    feature_names =  X.columns.tolist()
           
    feature_dict = dict((k,[]) for k in feature_names)
    feature_mean_dict = dict((k,[]) for k in feature_names)
    
    for i in range(X.shape[0]):
        print("%i/%i" % (i, X.shape[0]), end="\r")
        for c, feature in sorted(zip(contributions[i][:,1], feature_names), key=lambda x: -abs(x[0])):
            feature_dict[feature].append(round(c,2))
            
    for feature in feature_names:
        feature_mean_dict[feature] = np.mean(feature_dict[feature])
    
    features = list(feature_mean_dict.keys())
    values = list(feature_mean_dict.values())
    coef = pd.Series(values, index = features)
    imp_coef = coef.sort_values()
    plt.rcParams['figure.figsize'] = (10.0, 12.0)
    
    plt.figure()
    imp_coef[:50].plot(kind = "barh")
    plt.title("Bottom 50 Feature Contributions in Random Forest Model")
    
    plt.figure()
    imp_coef[-50:].plot(kind = "barh")
    plt.title("Top 50 Feature Contributions in Random Forest Model")
```

##### 2. Training function

```python
def run_classifiers(clfs, X_train, X_test, y_train, y_test):
    
    """ runs the training and the testing of the models """

    # kf = KFold(n_splits=10, shuffle=True, random_state=0)  KFold would be better.
    
    print("Number of test points: ", y_test.shape[0])
    print("\n")
    
    for i in clfs:
        
        # choice of classifier
        clf = clfs[i]
        
        # training
        start = time()
        clf.fit(X_train, y_train)
        time_fit = time() - start
        
        # testing
        start = time()
        y_pred = clf.predict(X_test)
        time_pred = time() - start
        
        # show metrics
        print(i)
        print("Training duration: %.2fs" % time_fit)
        print("Testing duration: %.2fs" % time_pred)
        print("\n")
        show_metrics(y_test, y_pred)
        print("\n")
        
        # for random forest only: show feature contribution and importance
        if i == 'RF':
            
            feature_names = list(X_test.columns)
            show_feature_importance(clf, X_test, feature_names)
            show_feature_contribution(clf, X_test)
```

### First model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

clfs = {
    'NBS': GaussianNB(),
    'CART': DecisionTreeClassifier(criterion = 'gini'),
    'RF': RandomForestClassifier(n_estimators = 50, random_state = 1)
}

run_classifiers(clfs, X_train_ohe_norm, X_test_ohe_norm, y_train_bin.ravel(), y_test_bin.ravel())
```
